```python
data1 = pd.read_csv("market_poll1.csv")
data2 = pd.read_csv("market_poll2.csv")

merged_data = pd.concat([data1, data2], axis=0) # объединили по строкам
```

```python
merged_data = merged_data.reset_index().drop(columns=['index'])
```

```python
merged_data['age'] = merged_data['age'][(merged_data['age']  >= 35) & (merged_data['age'] <=59)]
merged_data
```

```python
merged_data = merged_data.dropna()
```

```python
features = merged_data[['age', 'grade_service', 'grade_product']]
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()
```

```python
k = 2
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)

merged_data['cluster'] = cluster_labels
```

```python
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)

merged_data['cluster'] = cluster_labels

import seaborn as sns

sns.scatterplot(x='age', y='grade_service', hue='cluster', data=merged_data)
plt.title('Cluster Analysis')
plt.show()
```
